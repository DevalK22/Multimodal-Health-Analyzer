```python
import pandas as pd

df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
set = {}
```

```python
def process(d: str):
    d = d.capitalize
    return d
```

```python
for d in df['Disease']:
    set[process(d)] = True
```

```python
with open('file.txt', 'r+') as file:
    file.write(str(set.keys()))
```

```python
def process(s):
    s = s.strip()
    s = s.replace("_", " ")
    s = s.replace("  ", " ")
    return s
```

```python
symptoms = set()


for col in df.columns:
    if col == 'Disease':
        continue

    for s in df[col]:
        if type(s)==float:
            continue

        s = process(s)
        symptoms.add(s)
```

```python
print(symptoms.__str__())
```

```python
len(symptoms)
```

```python
mapds = {}
col = []

for _ in range(4920):
    col.append(0)

for s in symptoms: 
    # print(s)
    s = process(s)
    mapds[s] = col.copy()
    # print(mapds[s])

print(mapds.keys())

mapds['Diagnosis'] = []

# print(mapds)
```

```python
keys = []
for s in symptoms:
    keys.append(s)
print(keys)

with open('file.txt', 'r+') as file:
    file.write(keys.__str__())
```

```python
index = 0
for row in df.iterrows():
    for col in df.columns:
        if col=='Disease' or type(row[1][col])==float:
            continue            
        s = process(row[1][col])
        # print(row[1][col])
        mapds[s][index]=1
    mapds['Diagnosis'].append(row[1]['Disease'])
    index+=1
    
```

```python
newds = pd.DataFrame(mapds)
# newds.la
y_train = newds['Diagnosis'][0:4000]
xs = []
for s in symptoms:
    xs.append(s)
x_train = newds[xs][0:4000]

y_test = newds['Diagnosis'][4001:-1]
x_test = newds[xs][4001:-1]
```
